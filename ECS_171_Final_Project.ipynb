```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import arff
```

```python
data = arff.loadarff('rice.arff')
df = pd.DataFrame(data[0])
df.head()
```

```python
print(df.shape)
print(df.describe())
```

```python
df.plot(
    kind='box', 
    subplots=True, 
    sharey=False, 
    figsize=(25, 5)
)
plt.show()
```

```python
sns.pairplot(df, hue="Class", corner = True)
plt.show()
```

```python
x = df.drop('Class', axis = 1)
y = df['Class']
```

```python
sns.heatmap(x.corr(), vmin=-1, vmax=1, center=0, annot=True, cmap= 'RdBu')
plt.show()
```

```python
one_hot_encoded_data = pd.get_dummies(y)
print(one_hot_encoded_data)
```

```python
normalized_df=(x-x.min())/(x.max()-x.min())
print(normalized_df.head())
print(normalized_df.describe())
```
